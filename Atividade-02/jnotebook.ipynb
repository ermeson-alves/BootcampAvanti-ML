```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
imgBGR = cv2.imread('./Lenna.png')
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
imgRGB = cv2.resize(imgRGB, (500, 400))
r, g, b = cv2.split(imgRGB)

fig, axs = plt.subplots(1, 4, figsize=(18, 5))
axs[0].imshow(imgRGB) ; axs[0].set_title('Lenna RGB') ; axs[0].axis('off')
axs[1].imshow(r, cmap = 'gray') ; axs[1].set_title('Lenna R') ; axs[1].axis('off')
axs[2].imshow(g, cmap = 'gray') ; axs[2].set_title('Lenna G') ; axs[2].axis('off')
axs[3].imshow(b, cmap = 'gray') ; axs[3].set_title('Lenna B') ; axs[3].axis('off')


plt.show()
```

```python
print(f'Altura: {imgRGB.shape[0]}, Largura: {imgRGB.shape[1]}') # com numpy a tupla shape √© ordenada com base na ordem das dimens√µes, pelo que vi
print(f'N√∫mero de canais: {imgRGB.shape[2]}')
print(f'Tipo de dados: {imgRGB.dtype}')
print(f'Tom de cinza m√°ximo: {np.max(imgRGB)}')
print(f'Tom de cinza m√©dio: {np.mean(imgRGB):.2f}')
print(f'Tom de cinza m√≠nimo: {np.min(imgRGB)}')
```

```python
# Utilizarei a mesma imagem da quest√£o (1):
# Filtro da m√©dia
blur = cv2.blur(imgRGB,(3,3))

# Filtro da mediana
median = cv2.medianBlur(imgRGB,3)

# Filtro Gaussiano
gaussian = cv2.GaussianBlur(imgRGB, (3,3), 0)

fig, axs = plt.subplots(1, 4, figsize=(18, 8))
axs[0].imshow(imgRGB) ; axs[0].set_title('Lenna RGB') ; axs[0].axis('off')
axs[1].imshow(blur, cmap = 'gray') ; axs[1].set_title('Filtro da m√©dia') ; axs[1].axis('off')
axs[2].imshow(median, cmap = 'gray') ; axs[2].set_title('Filtro da mediana') ; axs[2].axis('off')
axs[3].imshow(gaussian, cmap = 'gray') ; axs[3].set_title('Filtro gaussiano') ; axs[3].axis('off')

plt.show()
```
